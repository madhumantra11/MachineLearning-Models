Preparing the data for training the model
We concatenate the PTRATIO and RM columns using np.c_ provided by the numpy library.

```python
X = pd.DataFrame(np.c_[bos['PTRATIO'], bos['RM']], columns = ['PTRATIO','RM'])
Y = bos['MEDV']
```

Splitting the data into Training and Test Data

We will now split the dataset into training and test data. We do this to train our model with 80% of the samples and test with the remaining 20%.

We are using the train_test_split function from the sklearn library to split the data.

```python
bos.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Training and testing the model
We use scikit-learnâ€™s LinearRegression to train our model on both the training and test sets.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```
